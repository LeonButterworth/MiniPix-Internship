```python
############################################################################
# Shows the total and average energy of alpha particles from its energy spectra
# plots these values
############################################################################

import numpy as np
from PIL import Image 
import pandas as pd
from matplotlib import pyplot as plt 

# open file
file_energy = open('C:\\Users\\Butter\\Desktop\\Distance Measurements\\3 cm\\Spectra\\One Frame\\Energy.txt')
Lines_energy = file_energy.read().splitlines()

particles = ['Alpha', 'Electron', 'Muon', 'Dot']

# append data to an array
energy_spectra=[] 
for line_energy in Lines_energy:
  tmpline_energy = line_energy.split('\t')
  energy_spectra.append([float(x) for x in np.array(tmpline_energy[:-1])])

# use spectra to find total amount of energy
total_energy = 0
for i in range(len(energy_spectra[0])):
    if energy_spectra[1][i] != 0: # checks whether there is a particle present in the spectra
      for j in range(int(energy_spectra[1][i])): # that energy is added based on the number of times it was measured
        total_energy += energy_spectra[0][i]

# print out values
print("Total energy =", total_energy)

total_number = sum(energy_spectra[1])
print("total number of alphas =", total_number)

avg_energy = int(round(total_energy / total_number))
print("Average energy =", avg_energy)

# plot spectra - only alphas
plt.figure(figsize=(10, 8), dpi=80)
for i in range(1):
  plt.plot(energy_spectra[2*i],energy_spectra[2*i+1], label=particles[i])

# print out the numbers for all the particles
for i in range(4):
  print(particles[i], "number =", sum(energy_spectra[2*i+1]))

plt.title("Energy spectrum")
plt.xlabel("Energy [keV]")
plt.ylabel("Particle counts")
plt.legend(loc="upper right")
plt.show()
```
